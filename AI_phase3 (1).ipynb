## ***Preprocessing***


We'll now drop the column ***CORPORATE_IDENTIFICATION_NUMBER*** since all its data is **unique** and the column ***REGISTERED_STATE*** since it has the the **same data** throughout.

```python
df=df.drop(['CORPORATE_IDENTIFICATION_NUMBER','REGISTERED_STATE'],axis=1)
```

```python
print(df['DATE_OF_REGISTRATION'].dropna().unique()[:10])
```

The dates in which the registration was made are now analyzed in the above code.

```python
df['DATE_OF_REGISTRATION'] = pd.to_datetime(df['DATE_OF_REGISTRATION'], format='%d-%m-%Y', errors='coerce')
```

The field ***DATE_OF_REGISTRATION*** is then converted to datetime format using the ***to_datetime*** function to assist in further preprocessing.

After this we'll now start handling the **missing values**.

```python
df.isnull().sum()
```

We first calculate the **total number of missing values** present in each column .

```python
missing_percentage = df.isnull().sum() * 100 / len(df)
print(missing_percentage)
```

This output shows the **percentage** of missing values present in each column.

```python
for col in df.select_dtypes(include=['float64', 'int64']):
    df[col].fillna(df[col].mean(), inplace=True)

for col in df.select_dtypes(include=['object']):
    df[col].fillna(df[col].mode()[0], inplace=True)
```

Next, the **missing values(nan)** in the ***numerical columns*** are replaced by their **mean** and those in the ***categorical column*** are replaced by their **mode**. After replacing, the dataframe will be as shown in the **output below**.

```python
df.head()
```

The **df.isnull().sum()** is used again to check if all the null values have been handled properly.

```python
df.isnull().sum()
```

## Exploratory Data Analysis

In this part of the preprocessing stage, graphs and plots are drawn to analyze and infer meaningful insights from the data.

Let us now visualize the ***annual count*** of registrations using a **line plot**.

```python
df.set_index('DATE_OF_REGISTRATION').resample('Y').size().plot()
```

From the above plot, we can infer that the **highest number** of registrations were made in the year ***2019***.

```python
values = df['COMPANY_STATUS'].value_counts()
labels = values.index

plt.bar(labels, values)
plt.title('Distribution of COMPANY_STATUS')
plt.xlabel('Status')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

The **bar plot** indicates that most registered companies in the RoC have the statuses '**ACTV**' and '**STOF**'.

```python
plt.scatter(df['AUTHORIZED_CAP'], df['PAIDUP_CAPITAL'], alpha=0.5)
plt.title('AUTHORIZED_CAP vs PAIDUP_CAPITAL')
plt.xlabel('AUTHORIZED_CAP')
plt.ylabel('PAIDUP_CAPITAL')
plt.show()
```

From the **scatter plot**, it's evident that '**AUTHORIZED_CAPITAL**' and '**PAIDUP_CAPITAL**' closely correlate. Both columns display 2-3 outliers, and an ***increase in one mirrors an ascent in the other***.

```python
plt.boxplot(df['AUTHORIZED_CAP'].dropna())
plt.title('Box Plot for AUTHORIZED_CAP')
plt.ylabel('Value')
plt.show()
```

We can see that the **thin box** plot shows little variation, and '***AUTHORIZED_CAPITAL***' has many outliers.

```python
time_data = df['DATE_OF_REGISTRATION'].value_counts().sort_index()
plt.plot(time_data.index, time_data.values)
plt.title('Registrations Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Registrations')
plt.show()
```

The plot depicts registrations from **1880** to **2019**, showing a significant rise starting in **1960** and continuing through **2019**.

```python
cross_tab = pd.crosstab(df['COMPANY_STATUS'], df['COMPANY_CLASS'])

plt.figure(figsize=(10, 8))
sns.heatmap(cross_tab, annot=True, cmap='Blues', fmt='g')
plt.title('Heatmap of COMPANY_STATUS vs. COMPANY_CLASS')
plt.show()
```

The **heatmap** depicts the relationship between **Company Status** and **Company Class**. Notably, most *PRIVATE CLASS* companies have either **ACTV** or **STOF** status

```python
sns.boxplot(x='COMPANY_STATUS', y='AUTHORIZED_CAP', data=df)
plt.title('Box plot of AUTHORIZED_CAP by COMPANY_STATUS')
```

The above plot shows us that companies with **ACTV** or **AMAL** status exhibit numerous outliers.

```python
sns.countplot(x='COMPANY_STATUS', hue='REGISTRAR_OF_COMPANIES', data=df)
plt.title('Distribution of COMPANY_STATUS by REGISTERED_STATE')
plt.xticks(rotation=45)
plt.legend(loc='upper right')
plt.show()
```

The **bar chart** illustrates company status distribution by count. ***ACTV*** and ***STOF*** dominate in numbers compared to others.

```python
df['YEAR_OF_REGISTRATION'] = df['DATE_OF_REGISTRATION'].dt.year
plt.figure(figsize=(20, 6))
sns.countplot(x='YEAR_OF_REGISTRATION', hue='COMPANY_CATEGORY', data=df)
plt.title('Distribution of COMPANY_CATEGORY over Years')
plt.xticks(rotation=90)
plt.legend(loc='upper right')
plt.show()
```

The plot displays *ROC* registrations from **1895** to **2019**. Registrations remained steady until the **1970s**, with noticeable increases beginning in **1995**.

## Encoding Categories

```python
for col in df.select_dtypes(include=['object']):
    print(col, df[col].nunique())
```

The above code outputs the **unique value counts** for each categorical column in df.

```python
for col in df.select_dtypes(include=['float64', 'int64']):
    df[col].fillna(0, inplace=True)
```

**Missing values** are filled with **0** for every **numeric column** in df.

```python
df=df.drop(['COMPANY_NAME'],axis=1)
```

***COMPANY_NAME*** column is **dropped** because most of its value are ***unique*** and it doesn't directly contribute to the final prediction.

```python
from sklearn.preprocessing import LabelEncoder
label_encoders = {}

for col in ['INDUSTRIAL_CLASS','PRINCIPAL_BUSINESS_ACTIVITY_AS_PER_CIN','EMAIL_ADDR','COMPANY_STATUS','COMPANY_CLASS','COMPANY_CATEGORY','COMPANY_SUB_CATEGORY','REGISTRAR_OF_COMPANIES']:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
    label_encoders[col] = le
```

The above code applies ***LabelEncoder*** to specified **categorical columns** in df and saves each encoder to label_encoders.

After this step, the categorical values get encoded and the dataframe will be as shown below.

```python
df.head()
```

```python
df=df.drop(['REGISTERED_OFFICE_ADDRESS','EMAIL_ADDR'],axis=1)
```

Since the columns ***REGISTERED_OFFICE_ADDRESS*** and ***EMAIL_ADDR*** don't contain data of any significance, they are ***dropped*** from the dataframe.The **dataset** after performing **categorical encoding** of data using Labelencoder and removing unnecessary columns will be as shown below:
