```python
from statsmodels.tsa.seasonal import seasonal_decompose

decomposition = seasonal_decompose(monthly_registrations, model='additive')
decomposition.plot()
plt.show()
```

```python
differenced_data = monthly_registrations.diff().dropna()
differenced_data
```

```python
from statsmodels.tsa.stattools import adfuller

result = adfuller(differenced_data)

print('ADF Statistic:', result[0])
print('p-value:', result[1])
print('Critical Values:', result[4])
```

```python
split_point = int(len(monthly_registrations) * 0.8)
train, validation = monthly_registrations[:split_point],monthly_registrations[split_point:]
```

```python
train
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(15, 6))
train.plot(label='Training Data')
validation.plot(label='Validation Data')
plt.title('Monthly Registrations')
plt.legend()
plt.show()
```

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX
model = SARIMAX(train, order=(1,1,1), seasonal_order=(1,1,1,12))
```

```python
results = model.fit()
```

```python
forecast = results.get_forecast(steps=len(validation)).predicted_mean
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np

mae = mean_absolute_error(validation, forecast)
rmse = np.sqrt(mean_squared_error(validation, forecast))

print(f"MAE: {mae}")
print(f"RMSE: {rmse}")

# Visualization
plt.figure(figsize=(15, 6))
train.plot(label='Training Data')
validation.plot(label='Validation Data')
forecast.plot(label='Forecast', color='orange')
plt.title('Monthly Registrations Forecast vs Actuals')
plt.legend()
plt.show()
```

```python
combined = train.append(validation)
final_model = SARIMAX(combined, order=(1,1,1), seasonal_order=(1, 1, 1, 12))
final_results = final_model.fit()
```

```python
future_forecast = final_results.get_forecast(steps=12).predicted_mean
print(future_forecast)
# Visualization
plt.figure(figsize=(15, 6))
combined.plot(label='Past Registrations')
future_forecast.plot(label='Future Forecast', color='orange')
plt.title('Monthly Registrations Future Forecast')
plt.legend()
plt.show()
```
